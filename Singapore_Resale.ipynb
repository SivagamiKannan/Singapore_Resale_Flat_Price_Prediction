# Outliers Handling- IQR Range

```python
df_final2=df_final1.copy()
```

# The Outlier Columns are
# floor_area_sqm
# storey_start
# sortey_end
# resale_price

```python
def outlier(df,column):
    q1=df[column].quantile(0.25)
    q3=df[column].quantile(0.75)
    iqr=q3-q1
    upper_threshold=q3+(1.5*iqr)
    lower_threshold=q1-(1.5*iqr)
    df[column]=df[column].clip(lower_threshold,upper_threshold)
```

```python
outlier_columns=["floor_area_sqm","storey_start","storey_end","resale_price"]
for j in outlier_columns:
    outlier(df_final2,j)
```

```python
for k in outlier_columns:
    plot(df_final2,k)
```

# Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
import pickle
```

```python
def model_selection(df,models):
    x=df.drop(columns=["resale_price"])
    y=df["resale_price"]
    #train test split
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
    model=models().fit(x_train,y_train)
    y_pred_train=model.predict(x_train)
    y_pred_test=model.predict(x_test)
    
    #checking the accuracy score
    mse_train=mean_squared_error(y_train,y_pred_train)
    rmse_train=np.sqrt(mse_train)
    mae_train=mean_absolute_error(y_train,y_pred_train)
    r2_train=r2_score(y_train,y_pred_train)

    mse_test=mean_squared_error(y_test,y_pred_test)
    rmse_test=np.sqrt(mse_test)
    mae_test=mean_absolute_error(y_test,y_pred_test)
    r2_test=r2_score(y_test,y_pred_test)
    
    train_metric={"Data":"Train Data",
           "R2_score":r2_train,
           "Mean_squared_error":mse_train,
            "Root_mean_squared_error":rmse_train,
            "Mean_absolute_error":mae_train,
            "Model":models
           }
    test_metric={"Data":"Test Data",
           "R2_score":r2_test,
           "Mean_squared_error":mse_test,
            "Root_mean_squared_error":rmse_test,
            "Mean_absolute_error":mae_test,
            "Model":models
           }
    print(train_metric)
    print(test_metric)
    print("")
```

```python
model_selection(df_final2, LinearRegression)
model_selection(df_final2, DecisionTreeRegressor)
model_selection(df_final2, RandomForestRegressor)
model_selection(df_final2, AdaBoostRegressor)
model_selection(df_final2, ExtraTreesRegressor)
model_selection(df_final2, GradientBoostingRegressor)
model_selection(df_final2, XGBRegressor)
```

```python
#Choosing randomforest regressor
x=df_final2.drop(columns=["resale_price"])
y=df_final2["resale_price"]

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model=RandomForestRegressor().fit(x_train,y_train)
y_pred_test=model.predict(x_test)

mse_test=mean_squared_error(y_test,y_pred_test)
rmse_test=np.sqrt(mse_test)
mae_test=mean_absolute_error(y_test,y_pred_test)
r2_test=r2_score(y_test,y_pred_test)

test_metrics={"Data": "Test Data",
        "Name":RandomForestRegressor.__name__,
        "R2_score":r2_test,
        "Mean_squared_error":mse_test,
        "Root_mean_squared_error":rmse_test,
        "Mean_absolute_error":mae_test
            }

print(test_metrics)
```
